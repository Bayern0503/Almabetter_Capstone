The Google Play Store, previously known as the Android Market, is a platform managed and developed by Google. It functions as the official app marketplace for certified devices running on Android and its related systems, as well as ChromeOS. Through this service, users can explore and download apps built using the Android SDK and uploaded by developers. Additionally, Google Play has offered digital media such as games, music, books, movies, and TV shows.


We were provided with two datasets: Play Store Data and User Reviews. Here's an overview of these files:


**Playstore Data.csv**: Contains detailed information about various apps available on Google Play. It includes 13 attributes describing each app.


**User Reviews.csv**: Contains up to 100 reviews per app, sorted by helpfulness. The text of these reviews has been pre-processed and enriched with three features: Sentiment (Positive, Negative, or Neutral), Sentiment Polarity, and Sentiment Subjectivity.

EDA involves summarizing the dataset with statistical methods and creating visualizations to extract meaningful insights and patterns. This process helps in better understanding the data and presenting it in an engaging way.

Steps in the EDA Process:

**Problem Definition:**

Begin by analyzing the dataset to understand the attributes provided and their relevance to the problem. This step involves exploring the data's context and its potential impact on the analysis.
Hypothesis Formation:

Based on the dataset’s attributes, formulate hypotheses that can guide the analysis. This helps identify potential trends, relationships, or patterns worth investigating.

**Data Cleaning:**

Address missing data, handle outliers, and prepare categorical variables for analysis to ensure the dataset is consistent and reliable.

**Univariate Analysis:**

Focus on individual attributes to understand their distribution and properties. This analysis is purely descriptive and aims to highlight patterns within single variables without considering relationships.

**Bivariate Analysis:**

Examine the relationship between two variables to understand how one might affect the other. This helps identify dependencies and correlations.


**Multivariate Analysis:**

Analyze interactions among three or more variables simultaneously to uncover complex relationships and trends.




**Hypothesis Testing:**

Validate the assumptions derived from the dataset using statistical methods to ensure the findings hold up under scrutiny.
EDA provides a foundation for exploring and interpreting the data effectively, ensuring valuable insights are derived from the analysis.

## 3. ***Data Wrangling***

### Data Wrangling Code

```python
# Write your code to make your dataset analysis ready.
#creating copies of original dataframe.
copy_ps_df = play_store_df.copy()
copy_ur_df = user_review_df.copy()

#dropping duplicates
copy_ps_df.drop_duplicates(keep='first',inplace=True)
copy_ur_df.drop_duplicates(keep='first',inplace=True)
```

```python
copy_ps_df.shape
```

```python
copy_ur_df.shape
```

```python
#dropping null values
copy_ps_df.isnull().sum()
copy_ps_df.dropna(inplace=True)
copy_ps_df.shape
```

```python
copy_ur_df.isnull().sum()
copy_ur_df.dropna(inplace=True)
copy_ur_df.shape
```

```python
copy_ps_df['Price'] = copy_ps_df['Price'].str.replace('$', '', regex=False)
copy_ps_df['Price'] = copy_ps_df['Price'].astype(float)
copy_ps_df['Reviews'] = copy_ps_df['Reviews'].astype(int)
copy_ps_df['Installs'] = copy_ps_df['Installs'].str.replace(',', '', regex=False)
copy_ps_df['Installs'] = copy_ps_df['Installs'].str.replace('+', '', regex=False)
copy_ps_df['Installs'] = copy_ps_df['Installs'].astype(int)
copy_ps_df['Size']= copy_ps_df['Size'].apply(lambda x: 'NaN' if x == 'Varies with device' else x)

def convert_to_numeric(value):
    if 'M' in value:
        return float(value.replace('M', '')) * 1_000_000
    elif 'k' in value:
        return float(value.replace('k', '')) * 1_000
    elif value == 'NaN':
        return None
    else:
        return float(value)

copy_ps_df['Size'] = copy_ps_df['Size'].apply(convert_to_numeric)


```

```python
# Category_Analysis
category_counts = copy_ps_df['Category'].value_counts()
plt.figure(figsize=(12, 6))
plt.bar(category_counts.index, category_counts.values)
plt.xlabel('Category')
plt.ylabel('Number of Apps')
plt.title('Number of Apps in Each Category')
plt.xticks(rotation=90)
plt.show()
```

```python
category_counts
```

```python
# Mean Rating
category_mean_ratings = copy_ps_df.groupby('Category')['Rating'].mean().sort_values()

# Calculate the overall mean rating
overall_mean_rating = copy_ps_df['Rating'].mean()

# Plotting
plt.figure(figsize=(12, 6))
sns.barplot(x=category_mean_ratings.index, y=category_mean_ratings.values, palette="viridis")
plt.axhline(y=overall_mean_rating, color='red', linestyle='--', label=f'Mean Rating: {overall_mean_rating:.2f}')

# Adding labels and title
plt.xticks(rotation=90)
plt.ylabel('Mean Rating')
plt.xlabel('Category')
plt.title('Mean Rating of Each Category vs Overall Mean Rating')
plt.legend()

# Show plot
plt.tight_layout()
plt.show()
```

```python

# Group by Genres and aggregate total installs
genre_installs = copy_ps_df.groupby('Genres')['Installs'].sum().reset_index()
genre_installs = genre_installs.sort_values(by='Installs', ascending=False)
genre_installs.head(10)
```

```python
# Top 10 genres by installs
top_genres = genre_installs.head(10)

# Plot bar chart
plt.bar(top_genres['Genres'], top_genres['Installs'], color='skyblue')
plt.title('Top Genres by Installs')
plt.xlabel('Genres')
plt.ylabel('Total Installs ')
plt.xticks(rotation=90)
plt.show()
```

```python
# Group by Genres and calculate total installs and average ratings
genre_analysis = copy_ps_df.groupby('Genres').agg({
    'Installs': 'sum',  # Total installs
    'Rating': 'mean'    # Average rating
}).reset_index()

# Sort by installs and ratings to find top performers
genre_analysis = genre_analysis.sort_values(by=['Installs', 'Rating'], ascending=[False, False])

# Display the top 10 genres
print(genre_analysis.head(10))
```

```python

# Scatter plot for all genres
plt.scatter(genre_analysis['Installs'], genre_analysis['Rating'], color='purple')
plt.title('Ratings vs Installs by Genre')
plt.xlabel('Total Installs')
plt.ylabel('Average Rating')
plt.grid(True)
plt.show()
```

```python
# Genre and Target Audience
# Group by Content_Rating and Genres and count occurrences
genre_audience = copy_ps_df.groupby(['Content_Rating', 'Genres']).size().reset_index(name='Count')

# Get top 3 genres for each Content_Rating
top_genres = genre_audience.groupby('Content_Rating').apply(lambda x: x.nlargest(3, 'Count')).reset_index(drop=True)

plt.figure(figsize=(12, 6))
sns.barplot(
    data=top_genres,
    x='Content_Rating',
    y='Count',
    hue='Genres',
    dodge=True
)

# Adding labels and title
plt.title('Top 3 Genres for Each Content Rating')
plt.xlabel('Content Rating')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Genres', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

plt.show()
```

```python
#Free vs Paid Apps
type_counts = copy_ps_df['Type'].value_counts()
plt.pie(
    type_counts,
    labels=type_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=['skyblue', 'orange']
)
```

```python
#Free to Paid app ratio in each category


# Group by Category and Type, and count the number of apps
free_paid_data = copy_ps_df.groupby(['Category', 'Type']).size().unstack(fill_value=0)

# Prepare data for the grouped bar chart
categories = free_paid_data.index
free_counts = free_paid_data['Free']
paid_counts = free_paid_data['Paid']

# Plot the grouped bar chart
x = np.arange(len(categories))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(20, 6))
bars1 = ax.bar(x - width/2, free_counts, width, label='Free Apps', color='skyblue')
bars2 = ax.bar(x + width/2, paid_counts, width, label='Paid Apps', color='orange')


ax.set_xlabel('Category')
ax.set_ylabel('Count of Apps')
ax.set_title('Count of Free and Paid Apps by Category')
ax.set_xticks(x)
ax.set_xticklabels(categories, rotation=45, ha='right')
ax.legend()

# Add data labels on the bars
for bar in bars1:
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height(), f'{bar.get_height()}', ha='center', va='bottom')
for bar in bars2:
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height(), f'{bar.get_height()}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
# Box plot to see the difference between the avg. rating of free and paid app.
plt.figure(figsize=(8, 6))
sns.boxplot(data=copy_ps_df, x='Type', y='Rating', palette='pastel')

plt.xlabel('App Type (Free or Paid)')
plt.ylabel('Rating')
plt.title('Effect of App Type (Free vs Paid) on Ratings')

plt.tight_layout()
plt.show()
```

```python
# Pair plot to see relation between Sentiment_subjectivity and Sentiment_Polarity
sns.pairplot(copy_ur_df, vars=['Sentiment_Polarity', 'Sentiment_Subjectivity'], hue='Sentiment')
plt.show()
```

Positive sentiments (blue) tend to have both higher polarity and higher subjectivity, indicating that positive reviews are typically highly opinion-based.

Negative sentiments (green) are less subjective overall but are concentrated in low polarity ranges.

Neutral sentiments (orange) cluster near zero polarity and low subjectivity, implying they are mostly factual or balanced.

There is a mild positive trend between Sentiment_Polarity and Sentiment_Subjectivity for positive and negative sentiments. Higher polarity reviews (more positive) also tend to be more subjective.
